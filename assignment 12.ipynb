```python
import pandas as pd
import numpy as np

# Creating a sample dataset for males and females
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
male_activities = {
    'Day': days,
    'Exercise_Time': np.random.randint(20, 120, size=7),  # Random exercise time for males
    'Work_Time': np.random.randint(240, 480, size=7),  # Random work time for males
    'Social_Media_Time': np.random.randint(0, 180, size=7),  # Random social media time for males
}
female_activities = {
    'Day': days,
    'Exercise_Time': np.random.randint(20, 120, size=7),  # Random exercise time for females
    'Work_Time': np.random.randint(240, 480, size=7),  # Random work time for females
    'Social_Media_Time': np.random.randint(0, 180, size=7),  # Random social media time for females
}

# Creating DataFrames for males and females
male_df = pd.DataFrame(male_activities)
female_df = pd.DataFrame(female_activities)

# Saving DataFrames to CSV files
male_df.to_csv('male_daily_life.csv', index=False)
female_df.to_csv('female_daily_life.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
male_df = pd.read_csv('male_daily_life.csv')
female_df = pd.read_csv('female_daily_life.csv')

# Histogram for Exercise Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.hist([male_df['Exercise_Time'], female_df['Exercise_Time']], bins=10, label=['Male', 'Female'], alpha=0.7)
plt.title('Exercise Time Comparison between Males and Females')
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for Work Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.bar(male_df['Day'], male_df['Work_Time'], label='Male Work Time', alpha=0.7)
plt.bar(female_df['Day'], female_df['Work_Time'], label='Female Work Time', alpha=0.7)
plt.title('Work Time Comparison between Males and Females')
plt.xlabel('Day')
plt.ylabel('Work Time (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
male_df = pd.read_csv('male_daily_life.csv')
female_df = pd.read_csv('female_daily_life.csv')

# Aggregate total social media time for males and females
male_social_media_total = male_df['Social_Media_Time'].sum()
female_social_media_total = female_df['Social_Media_Time'].sum()

# Pie chart for Social Media Time comparison between males and females
labels = ['Male', 'Female']
sizes = [male_social_media_total, female_social_media_total]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Social Media Time Comparison between Males and Females')
plt.show()
```
