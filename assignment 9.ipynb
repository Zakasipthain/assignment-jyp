```python
import csv
# Define purchase records
purchase_records = [
    {'Product': 'Grapes', 'Price': 10, 'Quantity': 5},
    {'Product': 'Banana', 'Price': 15, 'Quantity': 4},
    {'Product': 'Apple', 'Price': 20, 'Quantity':3 }
]
with open('purchase_records.csv', mode='w', newline='') as file:
    fieldnames = ['Product', 'Price', 'Quantity']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    for record in purchase_records:
        writer.writerow(record)


total_price = sum(record['Price'] * record['Quantity'] for record in purchase_records)

print("\nInvoice:")
print("------------------------------")
print("Product\t\tPrice\tQuantity")
print("------------------------------")
for record in purchase_records:
    print(f"{record['Product']}\t\t{record['Price']}\t{record['Quantity']}")
print("------------------------------")
print(f"Total Price: Rs {total_price}")
print("------------------------------")
```
