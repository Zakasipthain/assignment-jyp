```python
import numpy as np

# Let's use a different data set of exam scores
scores = np.array([72, 85, 90, 88, 82, 87, 94, 91, 76, 83])

# Calculate mean and median
mean = np.mean(scores)
median = np.median(scores)

# Compute standard deviation and variance
std_dev = np.std(scores)
variance = np.var(scores)

print(f"Mean: {mean}, Median: {median}, Standard Deviation: {std_dev}, Variance: {variance}")
```

```python
import matplotlib.pyplot as plt

scores = [72, 85, 90, 88, 82, 87, 94, 91, 76, 83]

# Define bins
bins = [0, 25, 50, 75,85, 100]

# Create histogram
plt.hist(scores, bins, edgecolor='black')

# Show histogram
plt.show()
```

```python
height = float(input("Enter your height in meters: "))
weight = float(input("Enter your weight in kilograms: "))

# Calculate the BMI using the formula
bmi = weight / (height ** 2)

# Print the BMI value
print("Your BMI is:", bmi)

# Display the health status based on the BMI range
if bmi < 18.5:
    print("You are underweight.")
elif bmi < 25:
    print("You are normal weight.")
elif bmi < 30:
    print("You are overweight.")
else:
    print("You are obese.")
```
